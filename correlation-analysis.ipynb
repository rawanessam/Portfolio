```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import ndtri
from sklearn.decomposition import PCA
df = pd.read_csv('assignemt 1 (2).csv')
```

```python
len_of_mem= df['Length of Membership']
len_of_mem1= len_of_mem.copy()
amount_spent= df['Yearly Amount Spent']
len_of_mem.corr(amount_spent, method = 'pearson') #original Correlation
```

```python
len_of_mem1.fillna(0, inplace = True)
len_of_mem1.corr(amount_spent, method = 'pearson') #filling empty cells with zeroes affects the correlation greatly 
```

```python
mean=len_of_mem.mean()
len_of_mem2= len_of_mem.copy()
len_of_mem2.fillna(mean,inplace=True)
len_of_mem2.corr(amount_spent, method = 'pearson') #filling empty cells with means does affect the correlation greatly
```

```python
median=len_of_mem.median()
len_of_mem3= len_of_mem.copy()
len_of_mem3.fillna(median,inplace=True)
len_of_mem3.corr(amount_spent, method = 'pearson')#replacing with median yields almost the same output as mean
#I would recommend filling gaps with mean or median
```

```python
len_of_mem.fillna(0, inplace = True)
amount_spent.fillna(0, inplace = True)
pca= PCA(n_components=2)
index= df.iloc[:,3:]
pca_res= pca.fit_transform(index)
plt.plot(pca_res)
```
